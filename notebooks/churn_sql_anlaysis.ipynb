## SQL QUERIES

```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel("data/telecom_churn.xlsx")
conn = sqlite3.connect("churn.db")
df.to_sql("customers", conn, if_exists="replace", index=False)
```

```python
query="""
SELECT * FROM customers
LIMIT 5
"""
result = pd.read_sql(query, conn)
print(result)
```

```python
query_1="""
SELECT count(*) AS TOTAL_CUSTOMERS FROM customers
"""
result=pd.read_sql(query_1,conn)
print(result)
```

```python
query_2= """
SELECT COUNT(*) AS churned_customers
FROM customers
WHERE Churn = 1;
"""
result = pd.read_sql(query_2, conn)
print(result)
```

```python
query_3= """
SELECT
(COUNT(CASE WHEN Churn = 1 THEN 1 END) * 100.0 / COUNT(*))
AS churn_rate
FROM customers;
"""
result = pd.read_sql(query_3, conn)
print(result)
```

```python
query_4= """
SELECT
Churn,
AVG(MonthlyCharge) AS avg_monthly_charge
FROM customers
GROUP BY Churn;
"""
result = pd.read_sql(query_4, conn)
print(result)
```

```python
query_5= """
SELECT
Churn,
AVG(CustServCalls) AS avg_service_calls
FROM customers
GROUP BY Churn;
"""
result = pd.read_sql(query_5, conn)
print(result)
```

```python
query_6= """
SELECT
ContractRenewal,
COUNT(*) AS total_customers,
SUM(CASE WHEN Churn = 1 THEN 1 ELSE 0 END) AS churned_customers
FROM customers
GROUP BY ContractRenewal;
"""
result = pd.read_sql(query_6, conn)
print(result)
```

```python
query_7= """
SELECT
Churn,
AVG(DataUsage) AS avg_data_usage
FROM customers
GROUP BY Churn;
"""
result = pd.read_sql(query_7, conn)
print(result)
```

```python
query_8= """
SELECT *
FROM customers
WHERE CustServCalls > 3
AND MonthlyCharge > 70;
"""
result = pd.read_sql(query_8, conn)
print(result)
```

```python
query_9 = """
SELECT
Churn,
AVG(RoamMins) AS avg_roaming_minutes
FROM customers
GROUP BY Churn;
          """
result = pd.read_sql(query_9, conn)
print(result)
```

## OVERALL INSIGHTS FROM QUERIES

1.Total Customer Base

The dataset contains 3,333 customers. This represents the total number of telecom users analyzed in the study.

Business meaning:
This dataset provides a moderate-sized sample to analyze customer behavior and identify churn patterns.

2.Customer Churn Rate

A portion of the customers have churned (left the telecom service). The churn rate represents the percentage of customers who discontinued the service.

Business meaning:
Understanding the churn rate helps telecom companies estimate revenue loss and identify retention opportunities.

3.Monthly Charges and Churn

Customers who churn tend to have higher average monthly charges compared to customers who stay.

Business insight:
Higher pricing may increase the likelihood of churn. Competitive pricing or targeted discounts may help reduce customer loss.

4.Customer Service Complaints

Customers who churn have more customer service calls on average.

Business insight:
Frequent complaints indicate dissatisfaction with the service. Improving customer support and resolving issues faster could reduce churn.

5.Contract Renewal Impact

Customers who did not renew their contracts show significantly higher churn compared to those who renewed.

Business insight:
Encouraging contract renewals through loyalty programs or incentives could improve retention.

6.Data Usage Patterns

Customers who churn show different data usage patterns compared to retained customers.

Business insight:
Understanding how customers use data can help telecom companies design better plans tailored to customer needs.

7.High-Risk Customers

Customers with More than 3 customer service calls and Monthly charges above 70 appear to be high-risk churn candidates.

Business insight:
These customers should be targeted with proactive retention strategies such as special offers or improved support.

8.Roaming Usage and Churn

Customers with higher roaming minutes show a slightly higher tendency to churn.

Business insight:
Roaming charges or service quality during travel might influence customer satisfaction.
